# Preparation

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
df.head()
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
df
```

```python
df.info()
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df,columns=['Geography', 'Gender'], drop_first = True)
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.drop('Exited', axis = 1)
y = df['Exited']
```

```python
X.head()
```

```python
y.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisasi. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}
from sklearn.linear_model import LogisticRegression
#START CODE
model1 = LogisticRegression()
params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid_LR = GridSearchCV(
             estimator= model1,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = -1, # core cpu yang digunakan
             cv = 5 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_LR.fit(X_train,y_train)
grid_LR.best_params_
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred_LR = grid_LR.predict(X_test)

print(classification_report(y_test, y_pred_LR))
print("")
print(confusion_matrix(y_test,y_pred_LR))
print("")
print(accuracy_score(y_test, y_pred_LR))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_LR)
plt.show()
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.neighbors import KNeighborsClassifier
#START CODE
model2 = KNeighborsClassifier()
params = {
    'n_neighbors' : [3, 5, 7],
    'weights' : ['uniform', 'distance'],
    'metric' : ['euclidean', 'manhattan']
}

#END CODE
from sklearn.model_selection import GridSearchCV
grid_KNN = GridSearchCV(
             estimator= model2,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = -1, # core cpu yang digunakan
             cv = 5 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_KNN.fit(X_train,y_train)
grid_KNN.best_params_
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred_KNN = grid_KNN.predict(X_test)

print(classification_report(y_test,y_pred_KNN))
print("")
print(confusion_matrix(y_test,y_pred_KNN))
print("")
print(accuracy_score(y_test, y_pred_KNN))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_KNN)
plt.show()
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}
from sklearn.ensemble import RandomForestClassifier

#START CODE
model3 = RandomForestClassifier()
params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, None],
    'max_features': ['sqrt', 'log2'],
    'min_samples_split': [2, 4, 6]
    }
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid_RF = GridSearchCV(
             estimator= model3,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = -1, # core cpu yang digunakan
             cv = 5 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid_RF.fit(X_train,y_train)
grid_RF.best_params_
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred_RF = grid_RF.predict(X_test)

print(classification_report(y_test,y_pred_RF))
print("")
print(confusion_matrix(y_test,y_pred_RF))
print("")
print(accuracy_score(y_test, y_pred_RF))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_RF)
plt.show()
```

## Tarik Kesimpulan Model Mana yang terbaik beserta alasannya

kesimuplannya adalah :
ada 3 model yang digunakan, yaitu Linear Regression, KNN dan Random Forest, lalu hasil accuracy yang terbaik adalah random forest. Mengapa random forest accuracynya terbaik? karena Random Forest memiliki Kemampuan Ensemble, adalah algoritma ensemble yang menggabungkan prediksi dari beberapa pohon keputusan. Ini dapat membantu mengurangi overfitting dan meningkatkan kinerja secara keseluruhan. Namun sebenarnya tidak ada model terbaik secara universal, tetapi yang terbaik akan bergantung pada konteks dan karakteristik pada sebuah datanya. Juga, ukuran dataset dan pengaturan parameter yang digunakan dapat berpengaruh besar pada hasil.
